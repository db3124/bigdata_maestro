```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func1(x):
    a = 2
    y = a * x ** 2
    
    return y

x = np.arange(-10, 10, 1)
y = func1(x)

plt.plot(x, y)

plt.axhline(y = 0, color = 'y', linestyle = ':')
plt.axvline(x = 0, color = 'y', linestyle = ':')

plt.show()
```
