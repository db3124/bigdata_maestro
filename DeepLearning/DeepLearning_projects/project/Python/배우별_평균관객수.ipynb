```python
import pandas as pd
import collections
import os
```

```python
df_movie = pd.read_excel('./병합_수정.xlsx')
df_movie
```

```python
# 영화 1회 찍은 배우

actor_list = []

for i in range(len(df_movie)):
    tmp_list = df_movie['배우'][i].split(',')
    
    for j in range(len(tmp_list)):
        actor_list.append(tmp_list[j])
        

tmp = dict(collections.Counter(actor_list))
tmp = sorted(tmp.items(), key=lambda x: x[1])

actor_name_once = []

for i in range(len(tmp)):
    if tmp[i][1] == 1:
        actor_name_once.append(tmp[i][0])

actor_name_once
```

```python
# 영화 1회 찍은 감독

director_list = []

for i in range(len(df_movie)):
    tmp_list = df_movie['감독'][i].split(',')
    
    for j in range(len(tmp_list)):
        director_list.append(tmp_list[j])
        

tmp = dict(collections.Counter(director_list))
tmp = sorted(tmp.items(), key=lambda x: x[1])

director_name_once = []

for i in range(len(tmp)):
    if tmp[i][1] == 1:
        director_name_once.append(tmp[i][0])

director_name_once
```

```python
# 영화 1회 배급한 배급사

distributor_list = []

for i in range(len(df_movie)):
    tmp_list = df_movie['배급사'][i].split(',')
    
    for j in range(len(tmp_list)):
        distributor_list.append(tmp_list[j])
        

tmp = dict(collections.Counter(distributor_list))
tmp = sorted(tmp.items(), key=lambda x: x[1])

distributor_name_once = []

for i in range(len(tmp)):
    if tmp[i][1] == 1:
        distributor_name_once.append(tmp[i][0])

distributor_name_once
```

```python
# 중복, 이상치 제거 후 리스트 변환
actor_list = list(set(actor_list))
actor_list.sort()
del(actor_list[0])
actor_list
```

```python
for i in range(len(df_movie)):
    df_movie['배우'][i] = ' ' + df_movie['배우'][i].replace(',', ' ') + ' '
    df_movie['감독'][i] = ' ' + df_movie['감독'][i].replace(',', ' ') + ' '
    
df_movie
```

```python
actor_audience_list = pd.DataFrame(columns=['배우', '평균관객수'])

for i in range(len(actor_list)):
    audience = 0 # 관객수 초기화 
    count = 0 # 영화 출연 횟수 초기화
    
    text = ' ' + actor_list[i] + ' '
    
    for j in range(len(df_movie)):
        if text in df_movie['배우'][j]:
            count += 1
            audience += df_movie['관객수'][j]
    
    actor_audience_list.loc[i, '배우'] = actor_list[i]
    actor_audience_list.loc[i, '평균관객수'] = audience / count
    
actor_audience_list = actor_audience_list.sort_values(by='평균관객수',ascending=False)
actor_audience_list
```

```python
actor_audience_list.to_excel('배우별_평균관객수.xlsx')
```

# # 배우별 평균점수를 반영한 영화점수

```python
# 배우_평균점수 추가 전처리

df_actor = pd.read_excel('./배우_점수.xlsx')

for i in range(len(df_actor)):
    if df_actor['배우'][i] in actor_name_once and df_actor['grade'][i] >= 70 :    
        df_actor['grade'][i] = df_actor['grade'][i] - 20
    
dir = os.getcwd() + './배우_평균점수2.xlsx'
writer = pd.ExcelWriter(dir, engine='xlsxwriter')
df_actor.to_excel(writer, 'w')
writer.save()
```

```python
tmp_actor = pd.DataFrame(columns=['영화', '배우점수'])

for i in range(len(df_movie)):
    num = 0
    score = 0
        
    for j in range(len(df_actor)):
        if df_actor['배우'][j] in df_movie['배우'][i]:
            num += 1 
            score += df_actor['grade'][j]
        
    tmp_actor.loc[i, '영화'] = df_movie['영화명'][i]
    tmp_actor.loc[i, '배우점수'] = score / num

tmp_actor
```
