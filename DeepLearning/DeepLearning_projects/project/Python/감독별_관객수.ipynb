```python
import pandas as pd
```

```python
df_movie = pd.read_excel('./파일/병합_수정.xlsx')
df_movie
```

```python
director_list = []

for i in range(len(df_movie)):
    tmp_list = df_movie['감독'][i].split(',')
    
    for j in range(len(tmp_list)):
        director_list.append(tmp_list[j])
```

```python
# 중복 제거
director_list = list(set(director_list))
director_list.sort()
director_list
```

```python
# 정확한 문자열 검색을 위해 앞뒤 공백 추가
for i in range(len(df_movie)):
    df_movie['감독'][i] = ' ' + df_movie['감독'][i].replace(',', ' ') + ' '
df_movie
```

```python
director_audience_list = pd.DataFrame(columns=['감독', '평균관객수'])

for i in range(len(director_list)):
    audience = 0 # 관객수 초기화
    count = 0 # 영화 제작 횟수 초기화
    
    text = ' ' + director_list[i] + ' '
    
    for j in range(len(df_movie)):
        if text in df_movie['감독'][j]:
            # count += 1
            audience += df_movie['관객수'][j]
    
    director_audience_list.loc[i, '감독'] = director_list[i]
    director_audience_list.loc[i, '평균관객수'] = audience # / count
    
director_audience_list = director_audience_list.sort_values(by='평균관객수',ascending=False)
director_audience_list
```

```python
# director_audience_list.to_excel('감독별_평균관객수.xlsx')
director_audience_list.to_excel('감독별_총관객수.xlsx')
```
