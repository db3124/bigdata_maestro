```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 

# seed값 설정
seed = 9
np.random.seed(seed)
tf.random.set_seed(seed)

df = pd.read_excel('./movie_score_re.xlsx')

dataset = df.values
X = dataset[:, 2:8]
Y = dataset[:, 1]

# 데이터 정규화
mean_x = X.mean(axis=0)
X_data = X - mean_x
std_x = X_data.std(axis=0)
X_data /= std_x

mean_y = Y.mean(axis=0)
Y_data = Y - mean_y
std_y = Y_data.std(axis=0)
Y_data /= std_y 

# train, test 데이터 분류
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.25, random_state=seed)

# 모델 생성
model = Sequential()
model.add(Dense(40, input_dim=6, activation='relu'))
model.add(Dense(26, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

# 모델 컴파일, 학습
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['MAE'])
model.fit(X_train, Y_train, epochs=300, batch_size=20)

Y_prediction = model.predict(X_test).flatten()

Y_pre = (Y_prediction * std_y) + mean_y
Y_t = (Y_test * std_y) + mean_y
```

# # 테스트 데이터 평가

```python
# 테스트 데이터
df_test = pd.read_excel('./201912_202001_개봉영화_최종.xlsx')

testset = df_test.values
test_X = testset[:, 2:].astype('float64')
test_Y = testset[:, 1].astype('float64')

X_data_test = test_X - mean_x
X_data_test /= std_x

Y_data_test = test_Y - mean_y
Y_data_test /= std_y
Y_prediction = model.predict(X_data_test).flatten()

Y_pre = Y_prediction * std_y + mean_y
Y_t = Y_data_test * std_y + mean_y

for i in range(len(Y_data_test)):
    label = '{:,}'.format(abs(int(Y_t[i])))
    prediction = '{:,}'.format(abs(int(Y_pre[i])))
    
    print("<%s> 실제관객 : %s 명 / 예상관객 : %s 명 "%(df_test['영화명'][i], label, prediction))
```

# # 관객수 예측 프로그램

```python
def main():
    
    actor_list = pd.read_excel('./배우_평균점수2.xlsx', index_col = 0)
    director_list = pd.read_excel('./감독_총점수.xlsx' )
    distributor_list = pd.read_excel('./배급사_평균점수.xlsx')

    for i in range(len(actor_list)):
        actor_list['배우'][i] = ' ' + actor_list['배우'][i] + ' '

    for i in range(len(director_list)):
        director_list['감독'][i] = ' ' + director_list['감독'][i] + ' '
        
        
    while True:
        try:
            actor = input('배우를 입력하세요 : ').split(',')

            actor_score = 0.0
            num = 0

            if len(actor) == 1:
                name = ' ' + actor[0] + ' '

                for i in range(len(actor_list)):
                    if actor_list['배우'][i] == name:
                        actor_score = actor_list['grade'][i]

            else:
                for index in range(len(actor)):

                    name = ' ' + actor[index] + ' '

                    for i in range(len(actor_list)):
                        if actor_list['배우'][i] == name:
                            num += 1
                            actor_score += actor_list['grade'][i]

                actor_score /= num

            print('배우점수는 %.2f입니다\n'%actor_score)
            break

        except:
            print('에러입니다. 구분자는 (,)를 이용하세요 : ')

    while True:
        try:
            director = input('감독을 입력하세요 : ').split(',')

            director_score = 0
            num = 0

            if len(director) == 1:
                name = ' ' + director[0] + ' '

                for i in range(len(director_list)):
                    if director_list['감독'][i] == name:
                        director_score = director_list['grade'][i]

            else:
                for index in range(len(director)):

                    name = ' ' + director[index] + ' '

                    for i in range(len(director_list)):
                        if director_list['감독'][i] == name:
                            num += 1
                            director_score += director_list['grade'][i]

                director_score /= num

            print('감독점수는 %.2f입니다\n'%director_score)
            break

        except:
            print('에러입니다. 구분자는 (,)를 이용하세요 : ')

    print('====================== 배급사 리스트 ======================')
    for name in distributor_list['배급사']:       
        print(name)
    print()

    while True:
        try:
            distributor = input('배급사를 입력하세요 : ').split(',')
            distributor_score = 0
            num = 0

            if len(distributor) == 1:
                name = distributor[0]

                for i in range(len(distributor_list['배급사'])):
                    if distributor_list['배급사'][i] == name:
                        distributor_score = distributor_list['grade'][i]

            else:
                for index in range(len(distributor)):

                    name = distributor[index]

                    for i in range(len(distributor_list)):
                        if distributor_list['배급사'][i] == name:
                            num += 1
                            distributor_score += distributor_list['grade'][i]

                distributor_score /= num

            print('배급사점수는 %.2f입니다\n'%distributor_score)
            break

        except:
            print('에러입니다. 구분자는 (,)를 이용하세요 : ')
    
    screen = input('스크린 수를 입력하세요 : ')
    time = input('상영 시간을 입력하세요 : ')
    age = input('관람연령을 입력하세요: ')
    
    input_X = np.array([[screen, time, actor_score, director_score, distributor_score, age]], 'float64')
    input_X

    input_X = input_X - mean_x
    input_X /= std_x

    Y_prediction = model.predict(input_X).flatten()
    Y_pre = int(abs(Y_prediction) * std_y + mean_y)
    Y_pre = '{:,}'.format(Y_pre)
    print("예상관객 : %s 명" % Y_pre)
```
