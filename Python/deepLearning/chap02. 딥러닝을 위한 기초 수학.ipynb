# 1. 1차함수 그리기

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# y = 2x + 3
def func1(x):
    a, b = 2, 3
    y = a * x + b
    
    return y 
```

```python
x = np.arange(-5, 5, 0.5)
y = func1(x)

print('x: ', x)
print('y: ', y)
```

```python
plt.plot(x, y)

# x축, y축 표시
plt.axhline(y = 0, color = 'k', linestyle = ':', linewidth = 1)
plt.axvline(x = 0, color = 'k', linestyle = ':', linewidth = 1)

plt.show()
```

# 2. 2차함수 그리기

```python
# def func2(x):
#     a = 0.5
#     y = a * x ** 2    
#     return y

def func2(x, p, q):
    a = 3
    y = a * (x - p) ** 2 + q
    return y
```

```python
x = np.arange(-50, 50, 0.1)
y = func2(x, 2, 3)

print('x: ', x)
print('y: ', y)
```

```python
plt.plot(x, y)

# x축, y축 표시
plt.axhline(y = 0, color = 'k', linestyle = ':', linewidth = 1)
plt.axvline(x = 0, color = 'k', linestyle = ':', linewidth = 1)

plt.show()
```

```python
plt.plot(x, y)

# x축, y축 표시
plt.axhline(y = 3, color = 'k', linestyle = ':', linewidth = 1)
plt.axvline(x = 2, color = 'k', linestyle = ':', linewidth = 1)

# x, y축 범위 지정
plt.xlim(-5, 10)
plt.ylim(0, 20)

plt.show()
```

# 3. 미분, 순간 변화율과 기울기

```python
def func3(x):
    return 0.01 * x ** 2 + 0.1 * x
```

```python
x = np.arange(0.0, 20.1, 0.1)
y = func3(x)
```

```python
plt.plot(x, y)

plt.show()
```

```python
def numerical_diff(f, x): # 함수의 매개변수로 함수를 넣는다. f = func3()
    h = 1e-4 # 델타x. 작은 수
    
    return (f(x + h) - f(x - h)) / (2 * h)
```

# 4. 시그모이드 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
#     return 1 / (1 + np.e ** (-x))

x = np.arange(-8, 8, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.axhline(y = 0, color = 'y', linestyle = ':', linewidth = 1)
plt.axhline(y = 1, color = 'y', linestyle = ':', linewidth = 1)
plt.axvline(x = 0, color = 'y', linestyle = ':', linewidth = 1)
plt.show()
```

# 5. 로그와 로그함수

```python
x = np.arange(0.01, 1, 0.01)

y1 = -np.log2(x)
y2 = -np.log(1 - x)

plt.plot(x, y1, x, y2)
plt.axhline(y = 0, color = 'y', linestyle = ':', linewidth = 1)
plt.axvline(x = 0, color = 'y', linestyle = ':', linewidth = 1)
plt.text(0.05, 6, 'y=-loga(x)', color = 'blue')
plt.text(0.8, 6, 'y=-loga(1-x)', color = 'brown')
plt.show()
```

```python
x = np.arange(-5, 5, 0.01)

y1 = np.log2(x)
y2 = 2 ** x

plt.plot(x, y1, x, y2)
plt.axhline(y = 0, color = 'y', linestyle = ':', linewidth = 1)
plt.axvline(x = 0, color = 'y', linestyle = ':', linewidth = 1)

plt.ylim(-30, 30)
plt.show()
```
