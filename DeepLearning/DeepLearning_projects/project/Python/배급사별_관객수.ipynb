```python
import pandas as pd
```

```python
df_movie = pd.read_excel('./파일/병합_수정.xlsx')
df_movie
```

```python
distributor_list = []

for i in range(len(df_movie)):
    tmp_list = df_movie['배급사'][i].split(',')
    
    for j in range(len(tmp_list)):
        distributor_list.append(tmp_list[j])
distributor_list
```

```python
# 리스트 내 중복 제거
distributor_list = list(set(distributor_list))
distributor_list.sort()
distributor_list
```

```python
distributor_audience_list = pd.DataFrame(columns=['배급사', '평균관객수'])

for i in range(len(distributor_list)):
    audience = 0 # 관객수 초기화 
    count = 0 # 배급사 영화 제작수 초기화
    
    text = distributor_list[i]
    
    for j in range(len(df_movie)):
        if text in df_movie['배급사'][j]:
            count += 1
            audience += df_movie['관객수'][j]
    
    distributor_audience_list.loc[i, '배급사'] = distributor_list[i]
    distributor_audience_list.loc[i, '평균관객수'] = audience / count
    
distributor_audience_list = distributor_audience_list.sort_values(by='평균관객수',ascending=False)
distributor_audience_list
```

```python
# distributor_audience_list.to_excel('배급사별_평균관객수.xlsx')
distributor_audience_list.to_excel('배급사별_평균관객수.xlsx')
```
